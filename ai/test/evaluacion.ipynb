```python
import pandas as pd
```

```python
question_answer = pd.read_csv('dataset_kantuta_group - dataset_kantuta_group.csv')
```

```python
question_answer.head(15)
```

```python
inputs = [ {"question": input_prompt} for input_prompt in question_answer['Consulta'] ]
outputs = [ {"answer": input_prompt} for input_prompt in question_answer['Respuesta esperada'] ]
```

```python
from langsmith import Client
from dotenv import load_dotenv
load_dotenv()
```

```python
client = Client()
dataset = client.create_dataset(
    dataset_name="Kantuta AI Evaluation Dataset v6",
    description="Dataset para evaluar el asistente virtual jurídico para Kantuta Group S.C."
)
```

```python

client.create_examples(inputs=inputs, outputs=outputs, dataset_id=dataset.id)
```
